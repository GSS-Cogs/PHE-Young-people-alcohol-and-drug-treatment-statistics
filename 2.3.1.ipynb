Table 2.3.1: Substance use of all young people in treatment 2016-17

```python
from gssutils import *

if is_interactive():
    scraper = Scraper('https://www.gov.uk/government/statistics/substance-misuse-treatment-for-young-people-statistics-2016-to-2017')
    tabs = {tab.name: tab for tab in scraper.distribution(title=lambda t: t.startswith('Data tables')).as_databaker()}
```

```python
tab = tabs['2.3.1 Substance Use']
```

```python
observations = tab.excel_ref('B5').expand(DOWN).expand(RIGHT).is_not_blank()
observations
```

```python
substance = tab.excel_ref('A5').expand(DOWN).is_not_blank()
substance
```

```python
sex = tab.excel_ref('B3').expand(RIGHT).is_not_blank() 
sex
```

```python
measuretype = tab.excel_ref('B4').expand(RIGHT).is_not_blank() 
measuretype
```

```python
Dimensions = [
            HDim(substance,'Substance',DIRECTLY,LEFT),
            HDim(sex,'Clients in treatment',CLOSEST,LEFT),
            HDim(measuretype,'Measure Type',DIRECTLY,ABOVE),
            HDimConst('Unit','People')            
            ]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
# if is_interactive():
#     savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
new_table = new_table[new_table['OBS'] != '' ]
```

```python
new_table = new_table[new_table['OBS'] != 0 ]
```

```python
new_table.columns = ['Value' if x=='OBS' else x for x in new_table.columns]
```

```python
new_table.head()
```

```python
new_table['Basis of treatment'] = 'Therapy'
```

```python
new_table['Substance'] = new_table['Substance'].map(
    lambda x: {
        'Total' : 'All' 
        }.get(x, x))
```

```python
new_table['Clients in treatment'] = new_table['Clients in treatment'].map(
    lambda x: {
        'Total' : 'All' 
        }.get(x, x))
```

```python
new_table.head()
```

```python
new_table['Measure Type'] = new_table['Measure Type'].map(
    lambda x: {
        'n' : 'Count', 
        '%' : 'Percentage',
        }.get(x, x))
```

```python
new_table.tail()
```

```python
new_table.dtypes
```

```python
new_table['Value'] = new_table['Value'].astype(str)
```

```python
new_table.head(3)
```

```python
new_table['Clients in treatment'] = new_table['Clients in treatment'].str.rstrip('^')
```

```python
new_table['Period'] = '2016-17'
new_table = new_table[['Period','Basis of treatment','Substance','Clients in treatment','Measure Type','Value','Unit']]
```
